```python
import os
from os import path
import shutil
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, dump, ElementTree
```

```python
copy = './sunny2/images/'
```

```python
import natsort

file_list = os.listdir('./sunny2/annotations/')
file_list = natsort.natsorted(file_list)
print(file_list)
```

```python
for i in range(len(os.listdir('./sunny2/annotations/'))):
    xml_file = './sunny2/annotations/sunny_'+str(i)+'.xml'
    tree = ET.parse(xml_file)
    root = tree.getroot()
    name_tag = root.findtext("file_name")
    origin = str(name_tag)
    shutil.copy(origin, copy)    
    
    dirname, basename = os.path.split(origin)
    origin_name = basename
    os.rename('./sunny2/images/'+str(origin_name), './sunny2/images/sunny_'+str(i)+'.png')
```
